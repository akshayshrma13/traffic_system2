```python
import numpy as np
from deepface import DeepFace

def get_face_embedding(image_path):
    """
    Detects a face, aligns it, and extracts a 512-dimensional embedding vector.
    Using 'Facenet512' for high accuracy and strong spatial performance.
    """
    try:
        # 'detector_backend' handles the initial bounding box tracking and alignment
        embedding_objs = DeepFace.represent(
            img_path=image_path, 
            model_name="Facenet512", 
            detector_backend="retinaface",  # High accuracy backend for surveillance/varying angles
            enforce_detection=True
        )
        # DeepFace returns a list of dictionaries (one per detected face)
        return embedding_objs[0]["embedding"]
    except Exception as e:
        print(f"Face detection failed for {image_path}: {e}")
        return None

def calculate_cosine_similarity(vector_a, vector_b):
    """
    Computes the cosine similarity between two normalized vector embeddings.
    Formula: (A . B) / (||A|| * ||B||)
    """
    dot_product = np.dot(vector_a, vector_b)
    norm_a = np.linalg.norm(vector_a)
    norm_b = np.linalg.norm(vector_b)
    return dot_product / (norm_a * norm_b)
```

```python
query_img_path = r"C:\Users\user\Desktop\Round2_flipkart_hackathon\face_database\satvik_query.png"
query_embedding = get_face_embedding(query_img_path)

if query_embedding is not None:
    # Simulated database mapping user labels to their registered profile embeddings
    database = {
        "User_A": get_face_embedding(r"C:\Users\user\Desktop\Round2_flipkart_hackathon\face_database\akshay_registered.jpg"),
        "User_B": get_face_embedding(r"C:\Users\user\Desktop\Round2_flipkart_hackathon\face_database\satvik_registered.jpg")
    }
    
    # Cosine threshold for FaceNet512 typically sits around 0.40 - 0.50
    SIMILARITY_THRESHOLD = 0.50  
    best_match = None
    highest_score = -1

    for user_name, registered_embedding in database.items():
        if registered_embedding is None: 
            continue
        
        score = calculate_cosine_similarity(query_embedding, registered_embedding)
        print(f"Similarity score with {user_name}: {score:.4f}")
        
        if score > highest_score:
            highest_score = score
            best_match = user_name

    # Check if verification passes threshold boundary conditions
    if highest_score >= SIMILARITY_THRESHOLD:
        print(f"\n[MATCH FOUND]: Identified as {best_match} (Score: {highest_score:.4f})")
    else:
        print("\n[UNKNOWN]: Face does not match any record in the database.")
```
